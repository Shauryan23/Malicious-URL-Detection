## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

## Importing the dataset

```python
dataset = pd.read_csv('malicious_url_dataset.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
def extract_year(arg):
    if(arg.item == 'None'):
        return
    else:
        return arg.str.slice(6, 10)
```

```python
dataset2 = dataset.apply(lambda x: extract_year(x) if(x.name == 'WHOIS_REGDATE') else x)
print(dataset2['WHOIS_REGDATE'])
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Feature Scaling

```python
# Scaling URL_LENGTH & NUMBER_SPECIAL_CHARACTERS

# Scaling X_train
X_train_features = X_train[:, 0:2]

scaler = StandardScaler().fit(X_train_features)
X_train_features = scaler.transform(X_train_features)

X_train[:, 0:2] = X_train_features

# Scaling X_test
X_test_features = X_test[:, 0:2]

scaler = StandardScaler().fit(X_test_features)
X_test_features = scaler.transform(X_test_features)

X_test[:, 0:2] = X_test_features
```
